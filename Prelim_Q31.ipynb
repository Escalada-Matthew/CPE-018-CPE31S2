```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
#2. Write a Python script using OpenCV that:
#Reads an image named "nature.jpg"

img = cv2.imread('/content/nature.jpg')
cv2_imshow(img)
```

```python
#Converts it to grayscale
gray = cv2.imread('/content/nature.jpg', 0)
cv2_imshow(gray)
```

```python
#Resizes it to 300x300 pixels
resized = cv2.resize(gray, (300, 300))
cv2_imshow(resized)
```

```python
#Saves the result as "processed.jpg"
cv2.imwrite("processed.jpg", img)
```
