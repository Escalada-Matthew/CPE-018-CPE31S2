```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
#You have an image named circles.png containing various circles  circles Links to an external site. of different sizes. Write a Python script using OpenCV that:
#Load and preprocess the image
#Detect circles using Hough Circle Transform
#Draw detected circles in red color
img = cv2.imread('/content/circles.png')
channel = img[:, :, 0]

blur = cv2.GaussianBlur(channel, (9,9), 1.5)
circles = cv2.HoughCircles(blur, cv2.HOUGH_GRADIENT, dp=1, minDist=150, param1=120, param2=45, minRadius= 10, maxRadius=0)
if circles is not None:
  circles = np.uint16(np.around(circles))
  for x, y, r in circles[0]:
    cv2.circle(img, (x, y), r, (0, 0, 255), 2)
    cv2.circle(img, (x, y), 2, (0, 0, 255), 3)
else:
  print("No circles found")

cv2_imshow(img)
```

```python
#Save the result as circles_detected.png
result = cv2.imwrite('circles_detected.png', img)
```
